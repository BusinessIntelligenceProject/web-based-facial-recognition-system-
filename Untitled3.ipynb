```python
import numpy as np
import cv2
import os
import dlib
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
```

```python
# Initialize lists to store image vectors and labels
X_data = []
y_labels = []

# Load the face detector and shape predictor
face_detector = cv2.CascadeClassifier("/content/drive/MyDrive/haarcascade_frontalface_default.xml")
!wget   http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2 # DOWNLOAD LINK
!bunzip2 /content/shape_predictor_68_face_landmarks.dat.bz2
datFile =  "/content/shape_predictor_68_face_landmarks.dat"
predictor = dlib.shape_predictor(datFile)
```

```python
# Initialize X_data as a 4-dimensional array
X_data = np.empty((0, 100, 100, 3))
for filename in os.listdir("/content/drive/MyDrive/Colab Notebooks/security_project/famous"):
  # Load image
  img_path = os.path.join("/content/drive/MyDrive/Colab Notebooks/security_project/famous", filename)
  img = cv2.imread(img_path)

  # Check if image is loaded successfully
  if img is not None:
    # Convert image to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_detector.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Iterate over detected faces
    for (x, y, w, h) in faces:
        # Extract the region of interest (ROI) containing the face
        face_roi = img[y:y+h, x:x+w]

        # Resize face ROI to a standard size (e.g., 100x100)
        face_roi = cv2.resize(face_roi, (100, 100))

        # Append the face ROI
        X_data = np.append(X_data, [face_roi], axis=0)
  else:
    print(f"Error loading image: {img_path}")

```

```python
# Convert lists to numpy arrays

y = np.array([0,2,0,1,2,1,1,0,0,0,0,0,0,5,5,5,5,5,5,5,2,2,2,2,2,2,2,2,1,0,2,0,1,1,1,1,1,1,1,4,4,4,4,4,4,3,3,3,3,3,3,3,2])
X_data = X_data[:len(y)]
X_data = np.array(X_data)
def create_cnn_model(input_shape, num_classes):
    model = models.Sequential([
        # Convolutional layers
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        # Flatten the feature maps
        layers.Flatten(),
        # Dense (fully connected) layers
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model
input_shape = (100, 100, 3)  # Assuming input images are resized to 100x100 RGB
num_classes = 6
cnn_model = create_cnn_model(input_shape, num_classes)
cnn_model.compile(optimizer='adam',
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])
X_train, X_test, y_train, y_test = train_test_split(X_data, y, test_size=0.15, random_state=42)

history = cnn_model.fit(X_train, y_train, epochs=20, validation_data=(X_test, y_test))
```

```python
# Convert lists to numpy arrays

fill = np.array(fill)

predictions = cnn_model.predict(fill)

# The predictions will be in the form of probabilities for each class
# You can get the predicted class by taking the index of the maximum probability for each prediction
predicted_labels = np.argmax(predictions, axis=1)

# Now, 'predicted_labels' will contain the predicted labels for your new data
print("Predicted Labels:", predictions)
class_names = ["assala", "mona zaki", "hend sabri", "saber rebai", "ragheb allama", "brad pitt"]
predicted_class_names = [class_names[label] for label in predicted_labels]

# Now, 'predicted_class_names' will contain the predicted class names for your new data
print("Predicted Class Names:", predicted_class_names)
```
